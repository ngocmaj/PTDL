```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv('Spotify.csv', encoding = 'iso-8859-1')
```

```python
df.head()
```

```python
genre_split = df['genre'].str.split(', ').explode()
genre_counts = genre_split.value_counts()
genre_counts
```

```python
genre_over_30 = genre_counts[genre_counts >= 30]
genre_over_30
```

```python
df_pred = pd.read_csv('SpotifyPrediction.csv', encoding = 'iso-8859-1')
```

```python
genre_split = df_pred['genre'].str.split(', ').explode()
genre_counts = genre_split.value_counts()
genre_counts
```

```python
df_pred['genre'] = df_pred['genre'].astype(str)
```

```python
df_pred['Pop'] = df_pred['genre'].apply(lambda x: 1 if 'Pop' in x else 0)

df_pred['Latin pop'] = df_pred['genre'].apply(lambda x: 1 if 'Latin pop' in x else 0)

df_pred['Hip hop'] = df_pred['genre'].apply(lambda x: 1 if 'Hip hop' in x else 0)

df_pred['Reggaeton'] = df_pred['genre'].apply(lambda x: 1 if 'Reggaeton' in x else 0)

df_pred['R&B'] = df_pred['genre'].apply(lambda x: 1 if 'R&B' in x else 0)

df_pred['Rap'] = df_pred['genre'].apply(lambda x: 1 if 'Rap' in x else 0)

df_pred['Rock'] = df_pred['genre'].apply(lambda x: 1 if 'Rock' in x else 0)

df_pred['K-Pop'] = df_pred['genre'].apply(lambda x: 1 if 'K-Pop' in x else 0)

df_pred['Soul'] = df_pred['genre'].apply(lambda x: 1 if 'Soul' in x else 0)

df_pred['Dance pop'] = df_pred['genre'].apply(lambda x: 1 if 'Dance pop' in x else 0)

df_pred['Others'] = df_pred['genre'].apply(lambda x: 1 if any(i for i in x.split(', ') if i not in ['Pop','Latin pop','Hip hop','Reggaeton','R&B','Rap','Rock','K-Pop','Soul','Dance pop']) else 0)
```
